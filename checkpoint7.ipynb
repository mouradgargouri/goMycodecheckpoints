Add a new column "Success" : if the score is higher than 10 we will have 1 else we will have 0

```python
A=[]
A= df['score'].tolist()
print(A)
B=[]
for i in A:
    if i>10:
        B.append(1)
    else:
         B.append(0)
print(B)
df['success'] = B
df
```
